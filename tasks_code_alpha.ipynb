```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Amazon Product Reviews Dataset
data = {
    "review": [
        "The product quality is amazing, I really loved it!",
        "Worst experience ever. Completely disappointed.",
        "It's okay, not too good and not too bad.",
        "Fast delivery and excellent packaging. Very satisfied!",
        "The item was damaged when it arrived."
    ]
}

df = pd.DataFrame(data)

# Define Positive & Negative Words
positive_words = ["amazing", "good", "love", "excellent", "fast", "satisfied", "great", "happy"]
negative_words = ["worst", "bad", "disappointed", "damaged", "poor", "late", "terrible"]

# Simple Sentiment Function
def simple_sentiment(text):
    text = text.lower()
    pos = sum(word in text for word in positive_words)
    neg = sum(word in text for word in negative_words)

    if pos > neg:
        return "Positive"
    elif neg > pos:
        return "Negative"
    else:
        return "Neutral"

# Apply Sentiment Analysis
df["sentiment"] = df["review"].apply(simple_sentiment)

# Count each Sentiment Type
sentiment_counts = df["sentiment"].value_counts()

# Plot Sentiment Distribution
plt.figure(figsize=(6, 4))
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Output the final table
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Restaurant": ["Aroma", "Spicy Hub", "Green Leaf", "Food Circle", "Taste Town", "Royal Biryani", "Pizza King", "Coffee Day"],
    "City": ["Chennai", "Chennai", "Madurai", "Coimbatore", "Chennai", "Madurai", "Coimbatore", "Chennai"],
    "Rating": [4.2, 3.8, 4.5, 4.1, 3.6, 4.3, 4.0, 3.9],
    "Price_Range": ["Medium", "Low", "Medium", "High", "Low", "High", "Medium", "Low"],
    "Online_Order": ["Yes", "Yes", "No", "Yes", "No", "Yes", "Yes", "Yes"]
}

df = pd.DataFrame(data)
print(df)

# Graph 1: Restaurants by City
plt.figure()
df['City'].value_counts().plot(kind='bar')
plt.title('Restaurants by City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()

# Graph 2: Rating Distribution
plt.figure()
df['Rating'].plot(kind='hist', bins=5, edgecolor='black')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
